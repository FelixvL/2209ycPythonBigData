```python
#top 10 sports
top_10_sport = data.sort_values(by=['Total'], ascending=False)
top_10_sport = top_10_sport.head(10)
top_10_sport
```

```python
#laatste 10 sports
laatste_10_sport = data.sort_values(by=['Total'], ascending=False)
laatste_10_sport = laatste_10_sport.tail(10)
laatste_10_sport
```

```python
#top 10 sports
sport_total_top = sns.lineplot(x='Total',
                               y='Sport', 
                               data=top_10_sport)
```

```python
#laatste 10 sports
sport_total_laatste = sns.lineplot(x='Total',
                          y='Sport', 
                          data=laatste_10_sport)
```

```python
#thema voor de plot
sns.set_theme(style='whitegrid')


#sport 
plot_sports = sns.barplot(data=top_10_sport,
                         x='Sport',
                         y='Power',
                         palette='dark',
                         alpha=.6)

#veranderen van de label namen
plot_sports.set(xlabel= None, 
                ylabel= 'Power')

#Rotatie voor de xticklabels
plot_sports.tick_params(axis='x', rotation=45)

#grootte van grafiek aanpassen

#plotten van grafiek
#fig.show()
```

```python
#welke vaardigheden worden het meest gebruikt

#Plotting the values for the satisfaction columns
plot_flex = sns.scatterplot(x=top_10_sport['Sport'],
                y=top_10_sport['Flexibility'])

#plot_house = sns.countplot(data['TevredenheidMetDeHuidigeWoning_1'], 
                           #ax=ax[1,0])

#verwijderen van de yticklabels
plot_flex.set(xlabel= None,
              ylabel= 'flexibility')

#Rotatie voor de xticklabels
plot_flex.tick_params(axis='x', rotation=45)


#fig.show()
```
